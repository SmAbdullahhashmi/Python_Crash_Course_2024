```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Plot
### A line plot is used to display data points connected by straight lines.

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.show()
```

## Scatter Plot
### A scatter plot is used to display individual data points without connecting lines.

```python
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

## Bar Plot
### A bar plot is used to represent categorical data with rectangular bars.

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 30, 15, 20]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

## Histogram
### A histogram is used to display the distribution of a dataset.

```python
data = [2, 3, 3, 3, 4, 4, 4, 5, 5, 6, 6, 6, 6, 7]

plt.hist(data, bins=5, edgecolor='k')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

## Pie Chart
### A pie chart is used to represent the parts of a whole.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 20, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```

## Box Plot
### A box plot is used to display the distribution and spread of a dataset.

```python
age = [15, 20, 25, 30, 35, 40, 45, 50]

plt.boxplot(age)
plt.xlabel('Box Plot')
plt.title('Box Plot Example')
plt.show()
```

## Heatmap
### A heatmap is used to visualize data in a matrix format.

```python
data = np.random.rand(5, 5)

plt.imshow(data, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()
```

How to read a correlation heatmap?

1. Look at the color of each cell to see the strength and direction of the correlation.                                        
2. Darker colors indicate stronger correlations, while lighter colors indicate weaker correlations.                             
3. Positive correlations (when one variable increases, the other variable tends to increase) are usually represented by warm colors, such as red or orange.                                                                                                 
4. Negative correlations (when one variable increases, the other variable tends to decrease) are usually represented by cool colors, such as blue or green.

```python
import seaborn as sns

# Load the Iris dataset
data = sns.load_dataset("iris")
```

```python
data
```

```python
# 2. Line Plot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Line Plot of Sepal Length vs Sepal Width")
plt.show()
```

```python
# 1. Scatter Plot
sns.scatterplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.show()
```

```python
# 3. Bar Plot
sns.barplot(x="species", y="sepal_length", data=data)
plt.title("Bar Plot of Sepal Length by Species")
plt.show()
```

```python
# 5. Box Plot
sns.boxplot(x="species", y="petal_length", data=data)
plt.title("Box Plot of Petal Length by Species")
plt.show()
```

A Violin Plot displays the following information:

A box plot that shows the median, quartiles, and potential outliers.
A rotated kernel density plot on each side, which provides a smoothed representation of the data distribution.
The width of the violin plot at any given point indicates the density or frequency of data points.

```python
# 6. Violin Plot
sns.violinplot(x="species", y="petal_length", data=data)
plt.title("Violin Plot of Petal Length by Species")
plt.show()
```

Pair Plot Respresentation:

Scatterplots: On the diagonal of the grid, you typically see histograms or kernel density plots for each variable in your dataset. These histograms show the distribution of each variable individually.

Off-diagonal Scatterplots: In the lower and upper triangles of the grid, you see scatterplots. Each scatterplot shows the relationship between two different variables. These scatterplots help you visualize how variables relate to each other.

```python
# 7. Pair Plot
sns.pairplot(data=data, hue="species")
plt.show()
```

```python
numeric_data = data.drop(columns="species")

# Create a correlation matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
